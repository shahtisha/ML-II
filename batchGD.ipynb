```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.5, 2.5])
y = np.array([0.2, 0.9])
w = 0
b = 0
alpha = 0.1

def batch_gd(x, y, w, b):
  epoch = 10
  weights = []
  errors = []
  for i in range(epoch):
    dw = 0
    db = 0
    error_epoch = 0
    for xi, yi in zip(x, y):
      dw += DW(xi, yi, w, b, alpha)
      db += DB(xi, yi, w, b, alpha)
      error_epoch += (yi - perceptron(xi, w, b))**2
    w += dw
    b += db
    errors.append(error_epoch / len(x))
    weights.append((w, b))
  print("final weights are ", w)
  print("final bias is ", b)
  return weights, errors
```

```python
def perceptron(x,w,b):
  y_in = np.dot(x, w) + b
  y_pred = sigmoid(y_in)
  return y_pred

def sigmoid(y_in):
  y_pred = 1 / (1 + np.exp(-y_in))
  return y_pred

def DW(x, y, w, b, alpha):
  y_pred = perceptron(x, w, b)
  dw = alpha*(y-y_pred)*(1-y_pred)*y_pred*x
  return dw

def DB(x, y, w, b, alpha):
  y_pred = perceptron(x, w, b)
  db = alpha*(y-y_pred)*(1-y_pred)*y_pred
  return db
```

```python
weights, errors = batch_gd(x, y, w, b)
```

```python
wt_0 = [wh[0] for wh in weights]
wt_1 = [wh[1] for wh in weights]

plt.figure(figsize=(13, 6))

plt.subplot(121)
plt.plot(wt_0)
plt.xlabel("epoch")
plt.ylabel("weight 0")


plt.subplot(122)
plt.plot(wt_1)
plt.xlabel("epoch")
```
