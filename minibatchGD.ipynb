```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron(x, w, b):
  y_in = np.dot(x, w)+b
  y_pred = sigmoid(y_in)
  return y_pred

def sigmoid(y_in):
  y_pred = 1/(1 + np.exp(-y_in))
  return y_pred

def DW(x, y, w, b):
  y_pred = perceptron(x, w, b)
  dw = (y-y_pred)*(1-y_pred)*y_pred*x
  return dw

def DB(x, y, w, b):
  y_pred = perceptron(x, w, b)
  db = (y-y_pred)*(1-y_pred)*y_pred
  return db
```

```python
def minibatch(x, y, w, b, alpha):
  epoch = 10
  errors = []
  error_epoch = 0
  batch_size = int(input("enter batch size: "))
  for i in range(epoch):
    dw, db, num =0,0,0
    for xi, yi in zip(x, y):
      dw += DW(xi, yi, w, b)
      db += DB(xi, yi, w, b)
      num += 1
      if num % batch_size == 0:
        w -= dw*alpha
        b -= db*alpha
      error_epoch += (yi - perceptron(xi, w, b))**2
    errors.append(error_epoch / len(x))
  print("squared error loss \n", errors)
  plt.plot(errors)
  plt.xlabel("epoch")
  plt.ylabel("loss")
  plt.show()
  return w, b

x = np.array([0.5, 2.5])
y = np.array([0.2, 0.9])
w = 0
b = 0
alpha = 0.1
new_w, new_b = minibatch(x, y, w, b, alpha)
print("updated weight is ", new_w)
print("updated bias is ",new_b)
```
